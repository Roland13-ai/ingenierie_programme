```python
## 1. Contexte théorique
Soit un circuit composé dune résistance $R$ et dun condensateur $C$ en série.
Limpédance équivalente est :
$$Z_{eq} = R + \frac{1}{jC\omega}$$
Le module et la phase sont :
$$|Z| = \sqrt{R^2 + \left(\frac{1}{C\omega}\right)^2}$$
$$\varphi = -\arctan\left(\frac{1}{RC\omega}\right)$$
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from pathlib import Path

# Chemin vers ton module circuits (jour3)
chemin_jour3 = Path.cwd().parent / "jour3"
sys.path.insert(0, str(chemin_jour3))
from circuits import impedance_resistance, impedance_condensateur, impedance_serie, module_et_phase

print("Modules importés avec succès.")
```

```python
# Paramètres
R = 1000      # 1 kΩ
C = 1e-6      # 1 µF
frequences = np.logspace(0, 5, 200)

# Calculs
modules = []
phases = []

for f in frequences:
    z_r = impedance_resistance(R)
    z_c = impedance_condensateur(C, f)
    z_eq = impedance_serie(z_r, z_c)
    mod, ph = module_et_phase(z_eq)
    modules.append(mod)
    phases.append(ph)

# Tracé
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.semilogx(frequences, modules)
ax1.set_ylabel("Module (Ω)")
ax1.set_title(f"Réponse en fréquence – RC série (R={R} Ω, C={C*1e6} µF)")
ax1.grid(True)

ax2.semilogx(frequences, phases)
ax2.set_xlabel("Fréquence (Hz)")
ax2.set_ylabel("Phase (°)")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
## 2. Interprétation

- *Module* : diminue avec la fréquence. Le circuit laisse passer les basses fréquences.
- *Phase* : tend vers -90° en hautes fréquences. Le comportement capacitif domine.

Ce circuit se comporte comme un *filtre passe-bas*.
    
```
